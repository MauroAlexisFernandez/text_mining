### Opcion 1 - Directamente con el trainer

#### Mimic creation of variables required:

````python
from huggingface_hub import notebook_login
notebook_login()
````

````python
def preprocess_function(examples):
    return tokenizer(examples['text'], padding="max_length", truncation=True)
````

````python
#NO-MODIFY: DATA LOAD
from datasets import load_dataset, Dataset, DatasetDict, ClassLabel
dataset = load_dataset("luisgasco/profner_classification_master")

````

````python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

# Cambia esto por tu modelo
hub_model_id = "luisgasco/modelo_test"

model = AutoModelForSequenceClassification.from_pretrained(hub_model_id)
tokenizer = AutoTokenizer.from_pretrained(hub_model_id)

dataset_tokenized = dataset.map(preprocess_function, batched=True)


training_args = TrainingArguments(
    output_dir="./results",
    per_device_eval_batch_size=32,
    do_train=False,
    do_eval=True,
    logging_dir="./logs",
    report_to="none"

#
  )
trainer = Trainer(
    model=model,
    args=training_args,
    eval_dataset=dataset_tokenized["validation"],
)
````

#### Evaluar en validation, predecir en test

En el procedimiento que hemos visto en clase, podr√≠amos lanzar la evaluaci√≥n utliizando el trainer:

````python
trainer.evaluate(dataset_tokenized["validation"])
````

**No** habr√≠a que lanzar esa evaluaci√≥n sobre `dataset_tokenize["test"]`, ya que no dispon√©is de datos etiquetados.
De hecho, la "label" del test set es siempre "-1", e incluso si intentamos hacer una predicci√≥n, no saltar√° un error.

````python
trainer.predict(dataset_tokenized["test"])
````

Para obtener las predicciones usando el trainer, hay que mapear el valor a 1 o 0 (los valores que el clasificador espera)

````python
dataset_tokenized["test"]["label"][0:4]
````

````python
def fix_labels(example):
    example["label"] = 1  # O lo que toque
    return example


# Aplica la funci√≥n al dataset de evaluaci√≥n
dataset_test = dataset_tokenized["test"].map(fix_labels)
````

````python
dataset_test["label"][0:4]
````

````python
predictions = trainer.predict(dataset_test)
````

Entonces, ya podemos procesar la salida y ponerlo en formato:

````python
y_pred = predictions.predictions.argmax(axis=1)
x_id = [x["tweet_id"] for x in dataset_test]
````

````python
import pandas as pd
output_df = pd.DataFrame(
    {'id': x_id,
     'label': y_pred
    })
````

````python
output_df.head(3)
````

````python
output_df.to_csv("APELLIDO1_APELLIDO2_NOMBRE_ejercicio1_predicciones.tsv", sep="\t",index=False)
````

### Opci√≥n 2:

Utilizando el m√©todo pipeline:

````python
from transformers import pipeline
import os
modelo_clasificacion = pipeline("text-classification", model="luisgasco/modelo_test", tokenizer="luisgasco/modelo_test")
````

````python
modelo_clasificacion
````

````python

label2id = {"SIN_PROFESION":0, "CON_PROFESION":1}

#NO-MODIFY: DATA LOAD
from datasets import load_dataset, Dataset, DatasetDict, ClassLabel
dataset = load_dataset("luisgasco/profner_classification_master")

dataset_test = dataset["test"]
````

````python
x_id = [x["tweet_id"] for x in dataset_test]
documentos = [x["text"] for x in dataset_test]
````

````python
documentos[0:3]
````

````python
from tqdm import tqdm
import torch
batch_size = 32
preds = []

for i in tqdm(range(0, len(documentos), batch_size)):
    batch = documentos[i:i+batch_size]
    with torch.no_grad():
        batch_preds = modelo_clasificacion(batch)
    preds.extend(batch_preds)
````

````python
preds[0:3]
````

````python
num_preds = [label2id[x["label"]] for x in preds]
````

````python
import pandas as pd
output_df2 = pd.DataFrame(
    {'id': x_id,
     'label': num_preds
    })
````

````python
output_df2.head(3)
````

````python
output_df2[["id","label"]].to_csv("APELLIDO1_APELLIDO2_NOMBRE_ejercicio1_predicciones2.tsv",index=False, sep="\t")
````
